## 8. Análisis de variables numéricas

Las variables numéricas permiten realizar análisis estadísticos sobre los datos. 
En este caso se analizan las coordenadas geográficas del dataset.

Durante el análisis se identificó que las columnas **LATITUD** y **LONGITUD**
estaban almacenadas como texto (object).

Para poder realizar cálculos estadísticos, se convirtieron a valores numéricos.

```python
df['LATITUD'].dtype
df['LATITUD'] = pd.to_numeric(df['LATITUD'], errors='coerce')
df['LONGITUD'] = pd.to_numeric(df['LONGITUD'], errors='coerce')
```

```python
df['LATITUD'].mean()
df['LATITUD'].median()
df['LATITUD'].min()
df['LATITUD'].max()

df[['LATITUD','LONGITUD']].isnull().sum()
```

## 9. Visualización de datos

Las visualizaciones permiten identificar patrones dentro del dataset.
A continuación se muestra un gráfico con las especies de árboles más frecuentes.

```python
top_species = df['ESPECIE'].value_counts().head(10)

top_species.plot(kind='bar', figsize=(10,5))

plt.title("Top 10 especies de árboles más registradas")
plt.xlabel("Especie")
plt.ylabel("Cantidad")

plt.show()
```
